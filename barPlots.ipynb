```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
%matplotlib inline
```

```python
nb_maps = 3
maps = ["Small", "Medium", "Large"]
toPlot = ["Computation Time", "Score", "# of Nodes"]
nb_toPlot = len(toPlot)
algo = ["DFS", "BFS", "Greedy Search", "A*"]
nb_algo = len(algo)
dfs = normalize(np.asarray([[0.02, 502, 15],
                            [0.233, 344, 306],
                            [0.599, 308, 521]]), axis=0)
dfs

bfs = normalize(np.asarray([[0.001, 502, 15],
                            [49.211, 570, 15878],
                            [1.467, 434, 1921]]), axis=0)
bfs
ucs = normalize(np.asarray([[0.0009, 502, 15],
                             [31.377, 570, 15878],
                             [0.978, 434, 1921]]), axis=0)

astar = normalize(np.asarray([[0.005, 502, 14],
                             [19.352, 570, 11246],
                             [1.411, 434, 1289]]), axis=0)
```

```python
plt.figure(figsize=(16, 10))
plt.title("Deep-First Search Algorithm over all layouts")
plt.xlabel("Maps")
plt.ylabel("Normalize statistics in relation with the Map")

bar_width = .5
bar_positions = np.arange(nb_maps)*bar_width*(nb_toPlot+1)

for i in range(nb_toPlot):
    plt.bar(bar_positions+bar_width*i, dfs[:, i], width=bar_width)

plt.xticks(bar_positions+(nb_algo-1)/2*bar_width, ["%s" % i for i in (maps)])
plt.legend(["%s" % i for i in (toPlot)])
plt.savefig("dfsBarPlots.png")
```

```python
plt.figure(figsize=(16, 10))
plt.title("Breadth-First Search Algorithm over all layouts")
plt.xlabel("Maps")
plt.ylabel("Normalize statistics in relation with the Map")

bar_width = .5
bar_positions = np.arange(nb_maps)*bar_width*(nb_toPlot+1)

for i in range(nb_toPlot):
    plt.bar(bar_positions+bar_width*i, bfs[:, i], width=bar_width)

plt.xticks(bar_positions+(nb_algo-1)/2*bar_width, ["%s" % i for i in (maps)])
plt.legend(["%s" % i for i in (toPlot)])
plt.savefig("bfsBarPlots.png")
```
